Lab 3: Error de truncamiento
Escoja una función, hacerle la expansión de Taylor alrededor de un punto  x0  hasta un orden N.

Dibujar la expasión para varios ordenes  N  en conjunto con la función según numpy.

¿Qué puede decir respecto al error de truncamiento?

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python

x0 = 0
n =3

x = sp.Symbol('x') # Me expresa simbolos
t = sp.exp(-x**2).series(x, x0, n).removeO()#corta la serie hasta n termino
t
########### CORRER ESTA LINEA PRIMERO##################
```

```python
t = sp.lambdify(x ,t) # convierte la funcion para poder manipularla
Array = np.linspace(x0 - 2, x0 + 2, 200)
Real = np.exp(-Array**2)   # función real
Taylor = t(Array)        # serie de Taylor truncada

# Gráfica
plt.plot(Array, Real, 'r-', label='e**(-x**2) con numpy')
plt.plot(Array, Taylor, 'b-', label="Serie de Taylor")
plt.legend()
plt.title('Serie de Taylor vs funcion real')
plt.grid(True)
plt.show()
```

la aproximación es muy buena alrededor de x0=0 pero el error de truncamiento crece rápidamente cuando nos alejamos de ese punto.

 aumentar el orden n la curva de la serie se ajusta mejor a la función real y el error de truncamiento disminuye.


 Se verificó que al aumentar el orden N de la serie de Taylor al rededor de x0=0
	 la aproximación se vuelve cada vez más exacta. Con valores altos de n como n=30 la curva de la serie es prácticamente igual a la función real en el rango analizado, lo que indica que el error de truncamiento se reduce mucho.
